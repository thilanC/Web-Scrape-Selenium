```python
def clean_text(input_text):
    valid_chars = set(
        "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789"
        ".,;!?\"'-()[]{}<>@#$%^&*_+=/\\|~`: "
        "\n\r\t")

    clean_text = ''.join(char for char in input_text if char in valid_chars)

    return clean_text
```

```python
folder_number = 0 #max folder number is 26
json_file_number = 0
json_file_link_number = 0
```

```python
parent_folder = "Sub_Sub_specilities"

folder_name_list = os.listdir(parent_folder)

for i in range (folder_number,len(folder_name_list)):

    folder_name = folder_name_list[i]

    folder_path = os.path.join(parent_folder, folder_name)

    if os.path.isdir(folder_path):
        print(f"Files in folder {i}.'{folder_name}':")
        print("/"*50)
        
        json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]

        print(f"Length of json files :{len(json_files)}")
        
        for j in range (json_file_number,len(json_files)):

            json_file = json_files[j]
            print(f"Start the {j}.{json_file} file links.")
            print("*"*50)
                  
            json_file_path = os.path.join(folder_path, json_file)
            with open(json_file_path, 'r') as f:
                json_data = json.load(f)

            print(f"Number of links in a json file :{len(json_data)}")
            print('-'*50)

            for l in range(json_file_link_number,len(json_data)):
                    
                Sub_speciality = json_data[l]

                parts = Sub_speciality.split("/")
                if Sub_speciality.count("/") == 2:
                            
                    Url = f"https://www.uptodate.com{Sub_speciality}"
                    print(f"{l}",".",f"https://www.uptodate.com{Sub_speciality}")
                    driver.get(f"https://www.uptodate.com{Sub_speciality}")
                    time.sleep(3)

                    article =  driver.find_element(By.ID, "topicText") 
                    links = HTML(article.get_attribute("innerHTML"))

                    links1=links.data
                    soup = BeautifulSoup(links1, "html.parser")

                    list1=[]
                    for link in soup.find_all("p"):
                        clean = clean_text(link.text)
                        list1.append(clean)

                    json_string = json.dumps(list1)

                    name_sub = name_from_url(Url)

                    folder_path2 = f'Text/{folder_name}'

                    if not os.path.exists(folder_path2):
                        os.makedirs(folder_path2)
                        print(f"Folder created at {folder_path2}")

                        with open(f"Text/{folder_name}/{name_sub}.json", "w") as f:
                                f.write(json_string)
                        print(f"Text/{folder_name}/{name_sub}.json")
                            
                    else:
                        print("Folder already exists.")
                        with open(f"Text/{folder_name}/{name_sub}.json", "w") as f:
                            f.write(json_string)
                        print(f"Text/{folder_name}/{name_sub}.json")
                        
            json_file_link_number = 0
        
print(f"{folder_name} link's data has been extracted and stored succesfully")
print("*"*50)
```
